```python
import pandas as pd

headers = ['Order ID', 'Order Date', 'Units', 'State/Province', 'Postal Code', 'Customer ID', 'Customer Lat', 'Customer Lng', 'Factory', 'Factory Lat', 'Factory Lng', 'Distance', 'rn']

df = pd.read_csv("calculated_distances.csv", names=headers, header=None, index_col=False)

df.head()
```

```python
import plotly.express as px

custom_scale = ['#FFD5C2', '#A88CB4', '#4A5899']

fig = px.scatter_geo(df,
                     lat='Customer Lat',
                     lon='Customer Lng',
                     size='Units',
                     hover_name='Postal Code',
                     hover_data=['Distance'],
                     color='Distance',
                     color_continuous_scale=custom_scale,
                     projection="albers usa",
                     title="Customer Reach and Shipments"
                    )

fig.update_layout(
    showlegend=True,
    font=dict(color='#24264B', family='Segoe UI'),
    geo=dict(showlakes=True, lakecolor='#E6F4FA')
)

fig.show()
```

```python
fig.write_image("customer_reach_map.png")
```
