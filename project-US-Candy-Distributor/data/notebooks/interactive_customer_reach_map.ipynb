```python
import pandas as pd

fig_path = os.path.join("project-US-Candy-Distributor", "data", "processed", "calculated_distances.csv")

df = pd.read_csv(fig_path, index_col=False)
df.head()
```

```python
import plotly.express as px

custom_scale = ['#FFD5C2', '#A88CB4', '#4A5899']

fig = px.scatter_geo(df,
                     lat='Customer_Lat',
                     lon='Customer_Lng',
                     size='Units',
                     hover_name='Postal_Code',
                     hover_data=['Distance_Miles'],
                     color='Distance_Miles',
                     color_continuous_scale=custom_scale,
                     labels={'Distance_Miles': 'Distance (miles)'},
                     projection="albers usa",
                     title="Customer Reach and Number of Shipments"
                    )

fig.update_layout(
    showlegend=True,
    font=dict(color='#24264B', family='Segoe UI'),
    geo=dict(showlakes=True, lakecolor='#E6F4FA')
)


fig.show()
```

```python
fig_path = os.path.join("project-US-Candy-Distributor", "reports", "figures", "customer_reach_map.png")
fig.write_image(fig_path)
```
