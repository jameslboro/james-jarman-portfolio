```python
from IPython.display import HTML, display

# List of hex color codes
colours = [
    "#FB8281", "#FE9666", "#F4D25A", "#B59525", "#4AC5BB", "#168980", "#0F5C55", "#3599B8", "#6A9FB0", "#8AD4EB", "#A4DDEE", "#4A5899", "#24264B", "#1B4D5C", "#B12955", "#A66999", "#B687AC", "#DFBFBF", "#A78F8F", "#706060", "#5F6B6D", "#5F6B6D", "#7F898A", "#293537", "#1C2325", "#BB4A4A", "#BD7150", "#7B4F71", "#28738A", "#559CAD", "#475052"
]

# Generate HTML to display colors
html_content = "<div>"
for colour in colours:
    html_content += f"""
    <div style="display: flex; align-items: center; margin-bottom: 5px;">
        <div style="width: 100px; height: 20px; background-color: {colour}; margin-right: 10px;"></div>
        <span>{colour}</span>
    </div>
    """
html_content += "</div>"

# Display the HTML
display(HTML(html_content))
```
